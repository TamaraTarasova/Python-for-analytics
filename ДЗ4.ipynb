## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(6, 4))
sns.lineplot(x=df['condition'], y=df['price'])
plt.title('Зависимость стоимости дома от его состояния')
plt.xlabel('Состояние дома по 5-ти бальной шкале')
plt.ylabel('Стоимость')
plt.grid();
```

```python
Вывод:Стоимость дома напрямую зависит от его состояния. Однако если состояние дома на уровне от3 до 4 то стоимость дома может не зависеть от показателя состояния.
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(x=df['view'], y=df['price'])
plt.title('Зависимость стоимости дома от вида')
plt.xlabel('Состояние дома по 4-х бальной шкале')
plt.ylabel('Стоимость');
```

```python
Вывод: Наблюдаем прямую зависимость цены от вида. Стоимость домов с хорошим видом растет быстрее
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['floors'], y=df['price'])
plt.title('Зависимость стоимости дома от его этажности')
plt.xlabel('Количество этажей')
plt.ylabel('Стоимость');
```

```python
Вывод: нет прямой зависимости стоимости дома от его этажности. Дома с 2 и 3 этажами стоят примерно одинаково, а вот стоимость домов где этажей 2,5 и 3,5 значительно выше.
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(x=df['sqft_living'], y=df['price'])
plt.title('Зависимость стоимости дома от его жилой площади')
plt.xlabel('Жилая площадь дома')
plt.ylabel('Стоимость');
```

```python
Вывод: В целом наблюдается зависимость стоимости домов от их жилой площади, однако зависимость эта не прямая а скочкообразная.
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(x=df['grade'], y=df['price'])
plt.title('Зависимость стоимости дома от его качества')
plt.xlabel('Качество дома по 13-ти бальной шкале')
plt.ylabel('Стоимость');
```

```python
Вывод: Зависимость стоимости дома от качества прямая, причем дома с высокоми показателями качества значительно дороже, в то время как дома уровня качества ниже среднего стоят примерно одинаково.
```
